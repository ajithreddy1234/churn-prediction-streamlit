```python
analyzer = BivariateAnalyzer(CategoricalVsNumericalAnalysis())
analyzer.execute_analysis(df, "Churn", "Tenure")
```

### Bivariate Analysis: Churn vs Tenure

The box plot compares **Tenure** across customers who churned and those who did not:

- Customers who **churned** generally have **shorter tenure**.
- The median tenure for churned customers is **significantly lower** compared to those who did not churn.
- There are many **outliers** among churned customers, indicating that some long-tenured customers still decided to leave.
- Non-churned customers tend to have a **higher and more consistent tenure**, with a wider interquartile range.

#### Insight:
Shorter tenure is associated with higher churn. This suggests that customers are more likely to leave early in their lifecycle, highlighting the importance of early engagement and retention efforts.


```python
analyzer = BivariateAnalyzer(CategoricalVsNumericalAnalysis())
analyzer.execute_analysis(df, "Churn", "Age")
```

### Bivariate Analysis: Churn vs Age

This box plot shows the **Age** of customers who churned and those who did not:

- The **average (median) age** is similar for both churned and non-churned customers.
- Both groups have a similar overall age range.
- Customers who churned have more **outliers**, especially at the lower and higher ends of the age scale.
- This means that some **very young or very old** customers are leaving, but most customers are in the middle age range.

#### Insight:
Age alone may not be a strong reason for churn, but outliers show that a few customers from extreme age groups are also churning. It may help when combined with other features.
